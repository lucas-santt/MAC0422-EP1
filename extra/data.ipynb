```python
import matplotlib.pyplot as plt
```

# Computer A

```python
sjf = [
    [0, 20], # esp
    [0, 30] # inesp
]

rr = [
    [74, 20],
    [83, 30]
]

prior = [
    [43, 26],
    [47, 30]
]
```

```python
# Para armazenar 2 gráficos em uma única imagem
plt.figure(figsize=(20, 6))
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))

ax1.bar(['SJF', 'Round Robin', 'Por Prioridade'], [sjf[1][1], rr[1][1], prior[1][1]], color=["#b4601f", "#e02609", "#ff0000"])
ax1.set_title('Deadlines concluídas em entrada-inesperado', fontweight='bold')
ax1.set_xlabel('Escalonador', fontweight='bold')
ax1.set_ylabel('Deadlines concluídas', fontweight='bold')
ax1.grid(axis='y')

ax2.bar(['SJF', 'Round Robin', 'Por Prioridade'], [sjf[0][1], rr[0][1], prior[0][1]], color=["#b4601f", "#e02609", "#ff0000"])
ax2.set_title('Deadlines concluídas em entrada-esperado', fontweight='bold')
ax2.set_xlabel('Escalonador', fontweight='bold')
ax2.set_ylabel('Deadlines concluídas', fontweight='bold')
ax2.grid(axis='y')
plt.show()
```

```python
plt.figure(figsize=(20, 6))
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))

ax1.bar(['SJF', 'Round Robin', 'Por Prioridade'], [sjf[0][0], rr[0][0], prior[0][0]], color=["#b4601f", "#e02609", "#ff0000"])
ax1.set_title('Preempções realizadas em entrada-inesperado', fontweight='bold')
ax1.set_xlabel('Escalonador', fontweight='bold')
ax1.set_ylabel('Preempções realizadas', fontweight='bold')
ax1.grid(axis='y')

ax2.bar(['SJF', 'Round Robin', 'Por Prioridade'], [sjf[1][0], rr[1][0], prior[1][0]], color=["#b4601f", "#e02609", "#ff0000"])
ax2.set_title('Preempções realizadas em entrada-esperado', fontweight='bold')
ax2.set_xlabel('Escalonador', fontweight='bold')
ax2.set_ylabel('Preempções realizadas', fontweight='bold')
ax2.set_ylim(0, 100) 
ax2.grid(axis='y')
plt.show()
```
